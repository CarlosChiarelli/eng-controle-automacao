````python
from numpy.linalg import inv, eig
from numpy.linalg import matrix_rank as rank
from numpy import array, eye, dot, zeros, poly
from numpy import roots, arange, asarray, concatenate
from functools import reduce
import matplotlib.pyplot as plt
from control import acker, place, ss, forced_response
from IPython.display import Image

def mul_mat(*args):
    return reduce(dot, args)

def plota_resp(T,
               yout,
               titulo,
               xlabel='Tempos (s)',
               ylabel='Altura da bolinha (m)\n'):

    plt.plot(T, yout)
    plt.title(titulo + '\n')
    plt.xlabel('\n' + xlabel)
    plt.ylabel(ylabel + '\n')
    plt.grid(alpha=.5)
    plt.show()
    
def plota_resp_sist(t, yout):
    
    plt.rcParams["figure.figsize"] = (11,7)
    
    fig, axes = plt.subplots(4, 1)
    ylabels = ["x1", "x1'", "x2", "x2'"]

    for num, ax in enumerate(axes.flatten()):
        ax.axhline(y=0, color='black')
        ax.plot(T, yout[num], color='b')
        ax.set_ylabel(ylabels[num])
        ax.grid()

    fig.tight_layout()
    plt.show()    
    
# plots config    
plt.rcParams["figure.figsize"] = (11,7)
plt.rcParams.update({'font.size': 15})
````

## Exerccio 01

Sistema eletro-mecnico para controlar a altura de uma bolinha de metal ao ar livre.
$$ $$

````python
Image('../img/07_06.png')
````

````python
Image('../img/07_07.png')
````

````python

A = array([[0,   1,    0],
           [980, 0, -2.8],
           [0,   0, -100]])

#B = array([0, 0, 100])
B = array([[0], [0], [100]])

C = array([1, 0, 0])

D, V = eig(A)

D.round(3)
````

````python

# SIMULANDO entrada = 0 e condies iniciais != 0
t = arange(0, 2, .01)
x0 = array([0.01, 0, 0])

sys = ss(A, B, C, 0)

T, yout, xout = forced_response(sys=sys, T=t, U=0, X0=x0)
````

````python

titulo = 'Entrada nula e condio inicial (altura bolinha) no zero'
plota_resp(T, yout, titulo)
````

Acima vemos que sem controle, sem fora imposta na bolinha soltando a bolinha numa pequena diferena da referncia, ela cai no cho.

Abaixo vamos determinar se o sistema  controlvel e observvel.
$$ $$

````python
P = array([B, mul_mat(A,B), mul_mat(A,A,B)]).T

P
````

````python
rank(P)
````

Rank cheio, logo o sistema  **controlvel**.
$$ $$

````python
# matriz de observabilidade
O = array([C, mul_mat(C,A), mul_mat(C,A,A)])

O
````

````python
rank(O)
````

Rank cheio, logo o sistema  **observvel**.

Em seguida ser alocado os polos desejados para o sistema.
$$ $$

````python
p1 = -10 + 10j
p2 = -10 - 10j
p3 = -50

# para polos multiplos preciso usar acker()
polos = array([p1, p2, p3])

K = place(A, B, polos)

K = asarray(K)

K
````

Acima vemos o ganho (controlador) do sistema.
$$ $$

````python

sys = ss(A-dot(B,K), B, C, 0)

T, yout, xout = forced_response(sys=sys, T=t, U=0, X0=x0)

titulo = 'Entrada nula e condio inicial (altura bolinha) no zero'
plota_resp(T, yout, titulo)
````

````python
p1 = -20 + 20j
p2 = -20 - 20j
p3 = -100

polos = array([p1, p2, p3])

K = place(A, B, polos)

K = asarray(K)

sys = ss(A-dot(B,K), B, C, 0)

T, yout, xout = forced_response(sys=sys, T=t, U=0, X0=x0)

titulo = 'Entrada nula e condio inicial (altura bolinha) no zero'
plota_resp(T, yout, titulo)

````

````python

T, yout, xout = forced_response(sys=sys, T=t, U=0.001)

titulo = 'Entrada no nula (0.001) e condio inicial zero'
plota_resp(T, yout, titulo)
````

$$ $$
Projetado o controlador, agora ser feito o projeto do obsevador.
$$ $$
**OBS**: os plos do observador precisam ser 5x maiores/mais rpidos que do modelo (controlador) de malha fechada. A referncia so os plos dominantes (com parte imaginria, prximos a esse eixo).
$$ $$

````python

p1 = -100
p2 = -101
p3 = -102

polos = array([p1, p2, p3])

# mudando estrutura da C (valores se mantm)
C2 = array([[1, 0, 0]])

L = place(A.T, C2.T, polos)

L = asarray(L)

L
````

$$ $$

Construo das matrizes do sistema (dinmica do observador).

$$ $$

```
At = array([[A - dot(B,K),       dot(B,K)],
            [zeros(A.shape), A - dot(L,C)]])

Bt = array([[B*Nbar        ],
            [zeros(B.shape)]])

Ct = array([C, zeros(C.shape)])
```
$$ $$

````python

# construindo At
a11 = A - dot(B,K)
a12 = dot(B,K)
a21 = zeros(A.shape)
a22 = A - dot(L.T,C2)

a1_ = concatenate((a11, a12), axis=1)
a2_ = concatenate((a21, a22), axis=1)

At = concatenate((a1_, a2_), axis=0)

# construindo Bt
b11 = B*Nbar
b21 = zeros(B.shape)

Bt = concatenate((b11, b21), axis=0)

# construindo Ct
c11 = C2
c12 = zeros(C2.shape)

Ct = concatenate((c11, c12), axis=1)
````

````python
x0_2 = concatenate((x0,x0), axis=0)

x0_2
````

````python

sys = ss(At, Bt, Ct, 0)

T, yout, xout = forced_response(sys=sys, T=t, U=0, X0=x0_2)

titulo = f'Entrada no nula (0.001) e condio inicial zero (com OBSERVADOR)'
plota_resp(T, yout, titulo)

````

````python
t = arange(0, 0.1, 1e-6)
ini_x0 = array([0.01, 0.5, -5])

x0_3 = concatenate((ini_x0,ini_x0), axis=0)

T, yout, xout = forced_response(sys=sys, T=t, U=0, X0=x0_3)

xout = xout.T
````

````python
n = 3
e = xout[:, n:]
x = xout[:, :n]
x_est = x-e
````

````python
xout.shape, e.shape, x.shape, x_est.shape
````

````python
h = x[:,0]
h_dot = x[:,1]
i = x[:,2]

h_est = x_est[:,0]
h_dot_est = x_est[:,1]
i_est = x_est[:,2]
````

````python
plt.plot(t,h,'-r', label='h')
plt.plot(t,h_est,':r', label='h_est')
plt.plot(t,h_dot,'-b', label='h_dot')
plt.plot(t,h_dot_est,':b', label='h_dot_est')
plt.plot(t,i,'-g', label='i')
plt.plot(t,i_est,':g', label='i_est')
plt.title("'Est'  o parmetro estimado\n")
plt.legend()
plt.grid(alpha=.4)
plt.show()
````
