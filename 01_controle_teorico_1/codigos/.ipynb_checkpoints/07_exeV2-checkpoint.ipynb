```python
Image('../dados/imagens/07_exe_1.png')
```

Apartir da função da tranferência temos que o polinômio característico é:

$$$$

$$ {s^4 + 2s^3 + 3s^2 + 4s + 5 = 0} $$

$$$$

```python
# raízes do polinômio
np.roots([1, 2, 3, 4, 5])
```

Aplicando a função roots, temos que raízes são:

$$$$

$$ {s_{1} = 0.2878 + 1.4261i} $$
$$ {s_{2} = 0.2878 - 1.4261i} $$
$$ {s_{3} = -1.2878 + 0.8579i} $$
$$ {s_{4} = -1.2878 - 0.8579i} $$

$$$$

Podemos observar que temos um par de polos com parte real positiva (SPD), por isso o sistema é instável.

```python
num = [1]
den = [1, 2, 3, 4, 5]
Gs  = (num, den)

LTI = lti(num, den)
t_step, y_step = step(LTI)
```

```python
plt.plot(t_step, y_step, label='y(t)', color='r')
plt.title('Resposta ao degrau unitário\n')
plt.xlabel('\n Tempo (s)')
plt.ylabel('y(t)\n')
plt.grid(color = 'b', alpha = 0.5, linestyle = 'dashed', linewidth = 0.5)
plt.show()
```

Apartir do gráfico gerado da resposta ao degrau, podemos observar e concluir que, de fato, o sistema é instável.

```python
s, K = sp.symbols('s, K')

G = (s+1)/(s*(s-1)*(s+6))
H = K*G/(1+K*G)
H
```

```python
K = 1
H = K*G/(1+K*G)
sp.factor(H, s)
```

$$ {s^3 + 5s^2 -5s + 1 = 0} $$

$$$$

```python
np.roots([1, 5, -5, 1])
```

$$$$

$$ {s_{1} = -5.8794} $$
$$ {s_{2} = 0.5921} $$

$$$$

Podemos observar dois polos positiva (SPD), por isso o sistema é instável.

```python
t = np.arange(0, 10, 1e-3)

# TF para k=1
num_k1 = [1,1]
den_k1 = [1,5,-5,1]

LTI        = lti(num_k1, den_k1)
t_k1, y_k1 = step(LTI, T=t)
```
