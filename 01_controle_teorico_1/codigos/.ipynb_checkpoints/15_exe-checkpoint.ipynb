```python
G1 = clt.tf([1], [0.01, 1])
G2 = clt.tf([1], [0.0001, 1])
```

```python
omega = np.arange(0, 1e6, 1)
M1, phi1, omega1 = clt.bode(G1, Plot=False, omega=omega)
M2, phi2, omega2 = clt.bode(G2, Plot=False, omega=omega)

db  = lambda x: 20*np.log10(x)
log = lambda x: np.log(x)
```

```python
plt.plot(log(omega1), db(M1), label='Gs1', color='g')
plt.plot(log(omega2), db(M2), label='Gs2', color='b')
plt.grid()
plt.legend()
plt.ylabel('Magnitude (dB)')
plt.xlabel('Frequência (rad/seg) \n[escala log]')
plt.title('Diagrama de Bode\n')
plt.show()
```

```python
plt.plot(log(omega1), phi1, label='Gs1', color='g')
plt.plot(log(omega2), phi2, label='Gs2', color='b')
plt.grid()
plt.legend()
plt.ylabel('Fase ()')
plt.xlabel('Frequência (rad/seg) \n[escala log]')
plt.title('Diagrama de Bode\n')
plt.show()
```

```python
x = clt.bode_plot(G1)
```

```python

from numpy import sin, pi

t = np.arange(0, 1, 1e-3)
u = lambda x: 50*sin(2*pi*10*x) + 100*sin(2*pi*100*x)
U = u(t)

tout, yout, x_out = clt.forced_response(G1, t, U)

plt.plot(t,    U,    label='Vi', color='b')
plt.plot(tout, yout, label='Vo', color='r')
plt.grid()
plt.legend()
plt.ylabel('v [V]')
plt.xlabel('tempo [s]')
plt.title('Resposta no tempo\n')
plt.show()
```
