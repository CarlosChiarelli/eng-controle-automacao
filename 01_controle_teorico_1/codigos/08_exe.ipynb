Apartir do DCL e do exercídio 5.1, temos:

$$ m\ddot{y}(t) + b\dot{y}(t) = F(t) $$

$$ $$

Aplicando a transformada de Laplace:

$$ $$

$$ ms^2Y(s) + bsY(s) = F(s) $$

$$ $$

Assim, a função de transferência é:

$$ $$

$$ G(s) = \frac{Y(s)}{F(s)} = \frac{1}{ms^2 + bs} $$

$$ $$

Fatorando a equação, temos:

$$ $$

$$ G(s) = \frac{1}{s(ms + b)} $$

$$ $$

Podemos descrever a planta como um sistema de primeira ordem acoplado a um sistema integrador $ \frac{1}{s} $, isto é, a função de transferência possui um polo $s = 0$. 

$$ $$

```python
def plotaResposta(t, saida, entrada, titulo, yLabel):
    # visulizando resposta
    plt.plot(t, saida,   label='y(t)', color='r')

    plt.title(titulo+' \n')
    plt.xlabel('\n Tempo (s)')
    plt.ylabel(yLabel +  '\n')
    plt.grid(color = 'b', alpha = 0.5, linestyle = 'dashed', linewidth = 0.5)
    plt.show()
```

```python
# vetor de tempo
dt = 1e-3
t  = np.arange(0, 20, dt)

# Parametros do modelo
m = 0.8
b = 0.4
```

```python
#Resposta ao degrau normalizada

tr  = np.arange(0, 20, dt) # tempo (0 a 20s)

numeradorFs = 1
denominadorFs = [m, b, 0]

Fs = (numeradorFs, denominadorFs)

LTI = lti(numeradorFs, denominadorFs) # gerando o sistema

t_step, y_step = step(LTI, T=tr)      # resposta ao degrau

r = np.ones(len(tr))                   

plotaResposta(t_step, y_step, r, 'Resposta ao degrau', 'Amplitude (m)')
```

# **Controle PID**

$$ K(s) = k_{p} + \frac{k_{i}}{s} + k_{d}s $$

$$$$

$$ H(s) = \frac{K(s)G(s)}{1 + K(s)G(s)} $$

$$$$

Inicialmento, consideramos $ k_{i} = k_{d} = 0 $ para ajustar $ k_{p} $ até atingir o requisito de tempo de subida.

$$ $$

Substituindo $ k_{i} = k_{d} = 0 $ e $ k_{p} = 1 $ em:

$$ $$

$$ K(s) = k_{p} + \frac{k_{i}}{s} + k_{d}s $$

$$ $$

Temos:

$$ K(s) = 1 $$

$$ $$

Assim:

$$ $$

$$ H(s) = \frac{K(s)G(s)}{1 + K(s)G(s)} = \frac{1}{ms^2 + bs} $$

$$ $$

```python
tr  = np.arange(0, 20, dt) # tempo (0 a 20s)
r = np.ones(len(tr))  

numeradorFs = 1
denominadorFs = [m, b, 1]

Fs = (numeradorFs, denominadorFs)

LTI = lti(numeradorFs, denominadorFs) # gerando o sistema

t_step, y_step = step(LTI, T=tr)      # resposta ao degrau                 

plotaResposta(t_step, y_step, r, 'Resposta ao degrau', 'Amplitude (m)')
```

```python
# Controle PD
Hs_num_PD = [0.15, 1]
Hs_den_PD = [0.008, 0.154, 1]

Hs_PD = (Hs_num_PD, Hs_den_PD)
```

```python

# entrada arbitrária
t = np.arange(0, 10, 1e-4)

def modela_entrada(t):
    if t>=1 and t<4:
        return 10
    elif t>=4 and t<6:
        return np.interp(t, [4,6], [10,4])
    elif t>=6 and t<8:
        return 4
    elif t>=8:
        return np.interp(t, [8,10], [4,0])
    else:
        return 0

u = list(map(modela_entrada, t))

# calculando respostas
t_PD,  y_PD,  x_out = lsim(Hs_PD, u, t)
t_PID, y_PID, x_out = lsim(Hs_PID, u, t)
```
