```python
# bibliotecas
import  pandas            as pd
import  numpy             as np
import  matplotlib.pyplot as plt
import  sympy             as sp
from    scipy             import integrate, optimize
from    numpy             import pi, sqrt, cos, sin
from    platform          import python_version
from    warnings          import filterwarnings
from    matplotlib.ticker import FormatStrFormatter
from    IPython.display   import Image

def mostra_valor(var, nomeVar, unidade, casasDec=None):
    
    printar = '\n%s = %.numCasasf [%s]\n'
    
    if casasDec:
        printar = printar.replace('numCasas', str(casasDec)) 
    else:
        printar = printar.replace('numCasas', str(2))
    
    print(printar % (nomeVar, var, unidade))

def cria_tabela():
    "{'Kf':np.nan, 'Kfs':np.nan, 'Se':np.nan, 'd':np.nan}"
    return pd.DataFrame({'Kf':[np.nan], 'Kfs':[np.nan], 'Se':[np.nan], 'd':[np.nan]})

def adiciona_linha(df, Kf, Kfs, Se, d):
    "Adiciona uma linha a tabela"
    ultima = df.shape[0]
    df.loc[ultima+1] = [Kf, Kfs, Se, d]
    return df
    
def obtem_diametros_df(d1_df, d2_df, d3_df):
    d1, d2, d3 = d1_df.loc[6,'d'], d2_df.loc[6,'d'], d3_df.loc[6,'d']
    return pd.DataFrame({'diametro':[1,2,3], 'valor [m]':[d1,d2,d3]})

def mostra_dados(dic, nome):
    ver = pd.DataFrame(dic)
    ver.index.name = nome
    return round(ver,1)

def plota_diagrama():
    plt.rcParams.update({'font.size': 16}) # tamanho da fonte

    titulos = ['Cisalhamento no plano Y-X', 'Cisalhamento no plano Z-X',
               'Momento no plano Y-X',      'Momento no plano Z-X',
               'Theta Y-X (graus)',         'Theta Y-Z (graus)']

    eixo_y = [cis_yx, cis_zx, M_yx, M_zx, theta_y, theta_z]
    cor    = ['b', 'r']*2 + ['g'] + ['g']
    x = np.linspace(0, L[-1], len(L))

    #fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(20, 10))
    fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(18, 12))

    # gerando e plotando 
    for titulo, y, cor, ax in zip(titulos, eixo_y, cor, axes.flat):

        ax.xaxis.set_major_formatter(FormatStrFormatter('%.2f'))

        ax.fill_between(x, y, color=cor, alpha=0.1)
        ax.plot(L, y, color=cor)

        ax.set_title(titulo)
        ax.hlines(0, 0, L[-1], colors='black', linewidth=3, linestyle = 'dashed')
        ax.grid(color = 'b', alpha = 0.5, linestyle = 'dashed', linewidth = 0.5)

    fig.tight_layout() # ajusta espaço entre gráficos
    plt.show()
    

filterwarnings('ignore')               # remove warnings
plt.rcParams.update({'font.size': 16}) # tamanho da fonte (gráficos)

print('\nVersão da Linguagem Python usada neste relatório:', python_version())
```

Primeiro é necessário calcular a deflexão pro mancal. Caso seja um ângulo acima de $0.04°$ é necessário um mancal que suporte deflexões nos eixo y e z.

```python
# escalonamento diâmetro
d_mancal = d_eixo[3] * 0.75
#d_mancal = np.floor(d_mancal*1e3) # [mm] (arrendonda pra baixo)
d_mancal = np.ceil(d_mancal*1e3) # [mm] (arrendonda pra baixo)

mostra_valor(d_mancal, 'Diâmetro mancal', 'mm')
```

```python

Fa, Fm = Rd['alt'], Rd['medio']

# força resultante
Fy = Fa['y'] + Fm['y']
Fz = Fa['z'] + Fm['z']
Fr = sqrt(Fy**2 + Fz**2)
Fa = 0

mostra_valor(Fr, 'Força radial resultante', 'N')
```

```python
# aco sae 1045
E = 200e9
#I_eixo = pi*(d_eixo[3]**4)/64

L = np.arange(0, A+B+C*1.1, 10e-4) # comprimento em X

def I_eixo(d):
    return pi*(d**4)/64
    
def V_y(x):
    if x < A:
        return Ra_y
    elif x >= A and x < (A+B):
        return Ra_y + Fg_y
    elif x >= (A+B) and x < (A+B+C):
        return Ra_y + Fg_y + Fp_y
    else:
        return Ra_y + Fg_y + Fp_y + Rd_y

def V_z(x):
    if x < A:
        return Ra_z
    elif x >= A and x < (A+B):
        return Ra_z + Fg_z
    elif x >= (A+B) and x < (A+B+C):
        return Ra_z + Fg_z + Fp_z
    else:
        return Ra_z + Fg_z + Fp_z + Rd_z

# deflexao viga
def M_z(x):
    if x < A:
        return Ra_z*x
    elif x >= A and x < (A+B):
        return Ra_z*x + Fg_z*(x-A)
    elif x >= (A+B) and x < (A+B+C):
        return Ra_z*x + Fg_z*(x-A) + Fp_z*(x-A-B)
    else:
        return Ra_z*x + Fg_z*(x-A) + Fp_z*(x-A-B) + Rd_z*(x-A-B-C)

def M_y(x):
    if x < A:
        return Ra_y*x
    elif x >= A and x < (A+B):
        return Ra_y*x + Fg_y*(x-A)
    elif x >= (A+B) and x < (A+B+C):
        return Ra_y*x + Fg_y*(x-A) + Fp_y*(x-A-B)
    else:
        return Ra_y*x + Fg_y*(x-A) + Fp_y*(x-A-B) + Rd_y*(x-A-B-C)

def theta(x, M_int):
    if x < A:
        return M_int/(I_eixo(d_eixo[1])*E)
    elif x >= A and x < (A+B):
        return M_int/(I_eixo(d_eixo[2])*E)
    elif x >= (A+B): # and x < (A+B+C):
        return M_int/(I_eixo(d_eixo[3])*E)
    
Fg_y, Fp_y = Fg['medio']['y']+Fg['alt']['y'], Fp['medio']['y']+Fp['alt']['y']
Ra_y, Rd_y = Ra['medio']['y']+Ra['alt']['y'], Rd['medio']['y']+Rd['alt']['y']

Fg_z, Fp_z = Fg['medio']['z']+Fg['alt']['z'], Fp['medio']['z']+Fp['alt']['z']
Ra_z, Rd_z = Ra['medio']['z']+Ra['alt']['z'], Rd['medio']['z']+Rd['alt']['z']
    
# esforço cortante (cisalhamento)
cis_yx = list(map(V_y, L))
cis_zx = list(map(V_z, L))

# momento fletor (flexão)
M_yx = list(map(M_y, L))
M_zx = list(map(M_z, L))

# theta (deflexão)
M_zx_int = integrate.cumtrapz(M_zx, L, initial=0)
M_yx_int = integrate.cumtrapz(M_yx, L, initial=0)

theta_z = [theta(l, m_int)*180/pi for l, m_int in zip(L, M_zx_int)]
theta_y = [theta(l, m_int)*180/pi for l, m_int in zip(L, M_yx_int)]

M_yx = list(map(M_y, L))
M_zx = list(map(M_z, L))
```

```python
n_linha = omega/60
U       = pi*d_mancal*n_linha

mostra_valor(n_linha, "n'", 'rps ou Hz')
mostra_valor(U,       'Velocidade tangencial', 'm/s')
```

$$$$

    * será adotado a relação **razão folga diametral/diâmetro** $= 0.0005$ como sugerido no *Norton* (livro texto).

```python
Cd = d_mancal*0.0005
#Cd = d_mancal*0.0017
#Cd = d_mancal*0.00017
#Cd = d_mancal*0.001
Cr = Cd/2

mostra_valor(Cd*1e6, 'Cd', 'micro metro')
```

$$$$
Obtenção do comprimento axial do mancal a partir da razão $\frac{L}{D}$.
$$$$

```python
l = d_mancal*0.5

mostra_valor(l*1e3, 'largura', 'mm')
```

```python
O_n = 30

epsilon_x = 0.21394 + 0.38517*np.log10(O_n) - 0.0008*(O_n - 60)

mostra_valor(epsilon_x, 'Epsilon_x', 'adimensional', 3)
```

```python
k_epsilon = O_n/(4*pi)

mostra_valor(k_epsilon, 'k_epsilon', 'adimensional')
```

```python
#visc_lub = (Nf*Fr*Cr**2)/(k_epsilon*U*l**3)
visc_lub = (Fr*Cr**2)/(k_epsilon*U*l**3)

mostra_valor(visc_lub, 'Viscosidade lubrificante (dinâmica)', 'Pa.s', 6)
mostra_valor(visc_lub*1e3, 'Viscosidade lubrificante (dinâmica)', 'cP', 4)
```

$$$$
O menor valor possível é para o óleo lubrificante **SAE ENGINE 50** obtendo uma viscosidade absoluta de $38.4$ **cP** para a temperatura de operação.

```python
############## pressão média filme de óleo ##############

p_med = Fr/(l*d_mancal) # Pa

mostra_valor(p_med*1e-6, 'Pressão média', 'MPa', 3)

# ângulo theta máx para pressão máxima
raiz_aux      = sqrt(1+24*epsilon_x**2)
theta_max_rad = np.arccos((1-raiz_aux)/(4*epsilon_x))
theta_max     = theta_max_rad * 180/pi

mostra_valor(theta_max, 'Theta máx (pressão máx)', 'graus')

############## pressão máxima ##############

visc_lub_oleo = visc_lub # Pa.s
z = 0
r = 0.5

num1 = (visc_lub_oleo*U)/(r*Cr**2) 
num2 = (l**2/4 - z**2)
num3 = (3*epsilon_x*sin(theta_max_rad))/(1 + epsilon_x*cos(theta_max_rad))**3

p_max = num1*num2*num3

mostra_valor(p_max*1e-6, 'Pressão máx', 'MPa')

############## ângulo de equilíbrio ##############

phi = np.arctan((pi*sqrt(1-epsilon_x**2))/(4*epsilon_x))

mostra_valor(phi*180/pi, 'Ângulo de equilíbrio', 'graus')

############## torque da parte estacionária ##############

num = visc_lub_oleo*(d_mancal**3)*l*n_linha*(pi**2)
den = Cd*(1-epsilon_x**2)**1/2
Ts  = num/den

mostra_valor(Ts, 'Torque estacionário', 'N.m', 4)

############## torque da parte rotacional ##############

exc = epsilon_x*Cr
Tr  = Ts + Fr*exc*sin(phi)  

mostra_valor(Tr, 'Torque rotacional', 'N.m', 4)

############## perda de potência no mancal ##############

perca_w = 2*pi*Tr*n_linha

mostra_valor(perca_w, 'Perca de potência mancal', 'J/s')

############## cálculo do coeficiente de atrito ##############

mi = 2*Tr/(Fr*d_mancal)

mostra_valor(mi, 'Coeficiente de atrito', 'adimensional', 3)
```

```python
h_min = Cr*(1-epsilon_x)

mostra_valor(h_min*1e7, 'Espessura mínima filme', 'E-7 m')
```

```python
# espessura do filme deve ser 4x menor

h_min2 = 4 * rug_med_sup

mostra_valor(h_min2*1e7, 'h_min', 'E-7 m')
```

```python
# Logo

epsilon = -(h_min2/Cr-1)

mostra_valor(epsilon, 'Epsilon', 'adimensional', 3)
```

```python
def myfunc(O_n):
    return 0.21394 + 0.38517*np.log10(O_n) - 0.0008*(O_n - 60) - epsilon

result = optimize.fsolve(myfunc, x0=40)
O_n    = result[0]

mostra_valor(O_n, 'Ocvirk', 'adimensional')
```

```python
k_epsilon = O_n/(4*pi)

mostra_valor(k_epsilon, 'k_epsilon', 'adimensional')
```

```python
P = (visc_lub*k_epsilon*U*l**3)/(Cr**2)
```

```python
Nf_proj = P/Fr

mostra_valor(Nf_proj, 'Coeficiente de segurança', 'adimensional')
```
