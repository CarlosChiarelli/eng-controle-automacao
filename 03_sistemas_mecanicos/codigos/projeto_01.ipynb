Calculandos as **componentes tangencias** e **radiais** das **forças** na **engrenagem** e **pinhão**:
$$ $$

```python
def forca_eng_pinhao(T):
    '''
    :param T: Float é o Torque
    :return:  2 Dictionarys (Fg, Fp) é o par das forças na engrenagem e pinhão (Fy, Fz)
    '''

    # forças engrenagem
    Fg_z  =  T/Re               # tangencial
    Fg_y  =  Fg_z * np.tan(phi) # radial
    Fg_z *= -1                  # correção sinal

    # forças pinhão
    Fp_z  = (T/Rp)             # tangencial
    Fp_y  = Fp_z * np.tan(phi) # radial
    Fp_z *= -1                 # correção sinal
    Fp_y *= -1                 # correção sinal

    return {'y':Fg_y,'z':Fg_z}, {'y':Fp_y,'z':Fp_z}
```

```python
Fg, Fp = {}, {} # Newton

Fg['medio'], Fp['medio'] = forca_eng_pinhao(T['medio'])
Fg['alt'],   Fp['alt']   = forca_eng_pinhao(T['alt'])
```

```python
def reacoes_mancais(Fg, Fp):
    '''
    :param Fg: Dictionary é a força na engrenagem {'y':,'z':}
    :param Fp: Dictionary é a força no pinhão     {'y':,'z':}
    :return:  2 Dictionarys (Ra, Rd) é o par das forças na engrenagem e pinhão (Fy, Fz)
    '''

    Fg_y, Fg_z = Fg['y'], Fg['z']
    Fp_y, Fp_z = Fp['y'], Fp['z']

    # SOMA Ma
    Rd_y = -1/(A+B+C)*(Fg_y*A + Fp_y*(A+B))
    Rd_z = -1/(A+B+C)*(Fg_z*A + Fp_z*(A+B))

    # SOMA Fz e Fy
    Ra_y = -(Rd_y + Fg_y + Fp_y)
    Ra_z = -(Rd_z + Fg_z + Fp_z)

    return {'y':Ra_y,'z':Ra_z}, {'y':Rd_y,'z':Rd_z}
```

```python
Ra, Rd = {}, {} # Newton

Ra['medio'], Rd['medio'] = reacoes_mancais(Fg['medio'], Fp['medio'])
Ra['alt'],   Rd['alt']   = reacoes_mancais(Fg['alt'],   Fp['alt'])
```

Obtemos as seguintes funções de singularidade:

$ $

$$ q = R_{a} \langle x-0 \rangle ^{-1} + F_{g} \langle x-0.05 \rangle ^{-1} + F_{p} \langle x-0.15 \rangle ^{-1} + R_{d} \langle x-0.2 \rangle ^{-1} $$

$ $

$$ V = R_{a} \langle x-0 \rangle ^0 + F_{g} \langle x-0.05 \rangle ^0 + F_{p} \langle x-0.15 \rangle ^0 + R_{d} \langle x-0.2 \rangle ^0 $$

$ $

$$ M = R_{a} \langle x-0 \rangle ^1 + F_{g} \langle x-0.05 \rangle ^1 + F_{p} \langle x-0.15 \rangle ^1 + R_{d} \langle x-0.2 \rangle ^1 $$

```python
L = np.arange(0, A+B+C*1.1, 10e-4) # comprimento em X

Fg_y, Fp_y, Ra_y, Rd_y = Fg['medio']['y'], Fp['medio']['y'], Ra['medio']['y'], Rd['medio']['y']
Fg_z, Fp_z, Ra_z, Rd_z = Fg['medio']['z'], Fp['medio']['z'], Ra['medio']['z'], Rd['medio']['z']

def V_y(x):
    if x < A:
        return Ra_y
    elif x >= A and x < (A+B):
        return Ra_y + Fg_y
    elif x >= (A+B) and x < (A+B+C):
        return Ra_y + Fg_y + Fp_y
    else:
        return Ra_y + Fg_y + Fp_y + Rd_y

def V_z(x):
    if x < A:
        return Ra_z
    elif x >= A and x < (A+B):
        return Ra_z + Fg_z
    elif x >= (A+B) and x < (A+B+C):
        return Ra_z + Fg_z + Fp_z
    else:
        return Ra_z + Fg_z + Fp_z + Rd_z

def M_z(x):
    if x < A:
        return Ra_z*x
    elif x >= A and x < (A+B):
        return Ra_z*x + Fg_z*(x-A)
    elif x >= (A+B) and x < (A+B+C):
        return Ra_z*x + Fg_z*(x-A) + Fp_z*(x-A-B)
    else:
        return Ra_z*x + Fg_z*(x-A) + Fp_z*(x-A-B) + Rd_z*(x-A-B-C)

def M_y(x):
    if x < A:
        return Ra_y*x
    elif x >= A and x < (A+B):
        return Ra_y*x + Fg_y*(x-A)
    elif x >= (A+B) and x < (A+B+C):
        return Ra_y*x + Fg_y*(x-A) + Fp_y*(x-A-B)
    else:
        return Ra_y*x + Fg_y*(x-A) + Fp_y*(x-A-B) + Rd_y*(x-A-B-C)

# esforço cortante (cisalhamento)
cis_yx = list(map(V_y, L))
cis_zx = list(map(V_z, L))

# momento fletor (flexão)
M_yx = list(map(M_y, L))
M_zx = list(map(M_z, L))

# torção (momento torsor)
T_torsao = np.zeros(len(L))
T_torsao[(L >= A) &(L < A+B)] = -T['medio']
```

## Processo iterativo

* chutar $K_{t}$ e $K_{ts}$ (em torno de $3$) e $C_{tam}$ (em torno de $1$)

* calcular $K_{f}$ e $K_{fs}$ (verificando $K_{fsm}$)

$$ $$
* tenho $d_{1}$, $d_{2}$, $d_{3}$
$$ $$
* tirar a do razão do entalhe pelos diâmetros
$$ $$
* $\frac{r}{d_{1}}$ e $\frac{r}{d_{3}}$
$$ $$
* razão sobre diâmetros ($\frac{D}{d}$), $D$ é sempre $d_{2}$, logo $\frac{d_{2}}{d_{1}}$ e $\frac{d_{2}}{d_{3}}$ 
$$ $$
* nos gráficos interpolar os valores pra achar $A$ e $b$ e achar $K_{t}$ e $K_{ts}$ e para chaveta obter os respectivos pontos no gráfico.
$$ $$
* nesse caso só temos 2, precisamos do $K_{t}$ e $K_{ts}$ pra seção do meio, pra isso é ncessário pegar o $K_{t}$ e $K_{ts}$ que for maior pra razão do $d_{1}$ e $d_{3}$
$$ $$
* calcular e iterar novamente até convergirem

```python
def calcula_Kt(d, A, b):
    " A*(r/d)**b "
    return A*(r/d)**b

def calcula_Kf(q, Kt):
    " 1 + q*(Kt-1) "
    return 1 + q*(Kt-1)

def calcula_d(Kf, Kfs, Se, Ma, Ta, Kfm, Kfsm, Sut, Mm, Tm):
    " Aplica fórmula extensa " 
    raiz_a = sqrt((Kf*Ma)**2  + 3/4*(Kfs*Ta)**2)
    raiz_b = sqrt((Kfm*Mm)**2 + 3/4*(Kfsm*Tm)**2)

    d = (32*Nf/pi*(raiz_a/Se + raiz_b/Sut))**(1/3)

    return d
```

## Chavetas

$$ $$

Agora será feito o projeto da chaveta. Será considerado chavetas quadradas com comprimento de $0.5$ $in$ seguido dos passos:

$$ $$
* calcular $F$ e $\tau$ (solicitações) nos pontos $B$ e $C$ (chaveta à esquerda e à direita)
$$ $$
* encontrar as tensões médias e máximas
$$ $$
* calcular o coeficiente de segurança ($N_{f}$)
$$ $$
* caso $N_{fCHAVETA}>=N_{fEIXO}$, redimensionar chaveta
$$ $$

Verificando a **Tabela 10-2** chegamos em larguras de $0.125$ $in$ para chaveta em $d_{1}$ e $d_{3}$, pois $0.875 < d <= 1.25$.

$$ $$

```python
# comprimento da chaveta 
L_chav = 0.5  *0.0254     # [m]
W_chav = 0.125*0.0254     # [m]
```

```python
# cálculo da resistência a fadiga corrigida

Se_linha = 0.5*Sut_chav # aço vida infinita
C_carreg = 1            # carregamento é flexão alternada

# C_tam
A95 = 0.05*W_chav*W_chav
d_eq = sqrt(A95/0.0766)
print('\nd_eq = %.2f mm\n' % (d_eq*1e3))
C_tam = 1

# material usinado (superfície) (mesmos exe anterior)
A_sup, b_sup  = 4.51, -0.265             
C_sup         = A_sup*((Sut_chav*1e-6)**b_sup) 

C_temp = 1     # T < 450°C
C_conf = 0.897 # 90%

Se_chav = C_carreg * C_tam * C_sup * C_temp * C_conf * Se_linha
print('Se_chav = %.1f MPa' % (Se_chav*1e-6))
```

$$ $$
## Ponto B (diâmetro 1)
$$ $$

```python
# forças na chaveta
Fa = T['alt']/(d1/2)
Fm = T['medio']/(d1/2)

# tensão de cisalhamento
Acis  = L_chav*W_chav      # área
tau_a = Fa/Acis
tau_m = Fm/Acis
```

```python
# tensão de Von Mises
tensao_a = sqrt(3*tau_a**2)
tensao_m = sqrt(3*tau_m**2)

Nf_chav1 = 1/(tensao_a/Se_chav + tensao_m/Sut_chav)

print('\nCoeficiente de segurança de falha por cisalhamento: %.2f\n' % Nf_chav1)
```

```python
# tensão de esmagamento
Aesm = (W_chav/2)*L_chav    # área de esmagamento (metade da largura)
tensao_max = (Fa+Fm)/Aesm   # tensão

Ns_chav1 = Sy_chav/tensao_max

print('\nCoeficiente de segurança por falha de esmagamento: %.2f\n' % Ns_chav1)
```

$$ $$
Acima vemos que a chaveta atendeu o necessário para esmagamento, contudo seu coeficiente de segurança para cisalhamento ainda está maior que o do eixo. Assim vamos diminuir com comprimento dela em $10$% para ver se atende o exigido.
$$ $$

```python
L_chav1 = L_chav*0.9

# forças na chaveta
Fa = T['alt']/(d1/2)
Fm = T['medio']/(d1/2)

# tensão de cisalhamento
Acis  = (L_chav1)*W_chav      # área
tau_a = Fa/Acis
tau_m = Fm/Acis

# tensão de Von Mises
tensao_a = sqrt(3*tau_a**2)
tensao_m = sqrt(3*tau_m**2)

Nf_chav1 = 1/(tensao_a/Se_chav + tensao_m/Sut_chav)

print('\nCoeficiente de segurança por falha de cisalhamento: %.2f\n' % Nf_chav1)
```
